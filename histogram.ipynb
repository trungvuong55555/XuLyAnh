```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('/content/pageBanner.png')
plt.imshow(img1)
plt.show()
```

```python
img1.shape
```

```python
#xu dung ham viet tay
vector = img1.ravel()
x= np.arange(0,256)
frequence = x
for i in x:
  count = 0
  for j in vector:
    if j==i:
      count = count+1
  frequence[i]= count
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121) , plt.plot(np.arange(0,256),frequence) ,plt.title("do thi xu dung ham viet tay")
plt.subplot(122) , plt.hist(vector,256,[0,256]), plt.title("Xu dung ham co san") # cac tham so lan luot la x, do to cua bins, range
plt.show()
```
